#In the given data we remove the Data Artchitechtin the job_title

```python
remove_job_title = jobs_df[jobs_df["job_title"] == "Data Architect"].index
remove_job_title
```

#We drop the remove_job_title

```python
jobs_subset = jobs_df.drop(remove_job_title)
jobs_subset
```

#We print the column data -- work_year, job_title, job_category, salary_currency, salary

```python
salary_job = jobs_subset[
    ["work_year", "job_title", "job_category", "salary_currency", "salary"]
]
salary_job
```

#We print the index of those salary, whose salary is greater than 130000

```python
remove_job_salary = salary_job[salary_job["salary"] >= 130000].index
remove_job_salary
```

#Print the remove salary

```python
short_jobs= salary_job.drop(remove_job_salary)
short_jobs
```

#Display the short_jobs.head(20)- Top 20

```python
display(short_jobs.head(20))
```

```python
colors = []
for index, row in salary_job.iterrows():
    job_title = row["job_title"]
    if job_title == "Data Scientist":
        colors.append("blue")
    elif job_title == "Data Analyst":
        colors.append("purple")
    elif job_title == "Research Analyst":
        colors.append("orange")
    else:
        colors.append("gray")
```

#We plot the graph Salary vs Salary Currency

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(
    salary_job["job_category"], salary_job["salary"], c=colors, marker="o"
)
plt.title("Salary in Year(s)")
plt.xlabel("Salary Currency")
plt.ylabel("Salary")
plt.show()
```
